```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv(r"C:\Users\LENOVO\Desktop\machine learnig\SVM\data.csv")
```

```python
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[0::,2:32]
y=df["diagnosis"]
le=LabelEncoder()
y_encoded=le.fit_transform(y)
```

```python
for i in range(2,32):
    plt.scatter(y,df.iloc[0::,i])
    plt.show()
```

```python
#standardizing the data 
X_train,X_test,y_train,y_test=train_test_split(X,y_encoded,test_size=0.3,random_state=42)
sd=StandardScaler()
sd.fit(X_train)
X_train1=sd.transform(X_train)
X_updated_train=pd.DataFrame(X_train1,columns=X_train.columns)
X_test1=sd.transform(X_test)
x_updated_test=pd.DataFrame(X_test1,columns=X_test.columns)
```

```python
svm=SVC()
svm.fit(X_updated_train,y_train)
```

```python
y_pred=svm.predict(x_updated_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,recall_score,f1_score
print("ACURRACY=",accuracy_score(y_test,y_pred))
print("classification_report",classification_report(y_test,y_pred))
print("confusion matrix ",confusion_matrix(y_test,y_pred) )
```
